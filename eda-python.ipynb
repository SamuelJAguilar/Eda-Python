# Analisis del Reporte EDA

**primeramente vamos analizar las alertas**

-El gameid presenta la alerta de Unique y Uniform, ya que el id va de 1 en 1 y son valores unicos, esta columna se puede eliminar ya que para el analisis de los datos no nos ayuda para hacer predicciones simplemente es una etiqueta para la correlacion de los datos

-La columna assists tiene ceros, sin embargo este no es un error y si representa un valor importante, ya que en esas partidas no ayudo a sus compañeros a realizar bajas en el juego, en el cual es un punto a que el jugador debe mejorar y se debe predecir como se obtiene esos resultados para que mejore

-La columna avg_dmg_delta tiene ceros esta columna es un promedio de daño que realizo el jugador en la partida, pero para que sea 0 es porque no se jugo la partida, ya sea porque el perdio la conexion o otra persona se salio y se rindieron sin jugar la partida, y el juego lo toma como empate si se rinden en la segunda ronda o como derrota si se rindieron en una ronda superior, por ende al no jugar la partida completa no se puede hacer un analisis adecuado de las demas estadisticas ya que el no jugo y no realizo nada en la partida, entonces seria de analizar la columna y ver cual fue el motivo por el cual el tiene 0 de daño, si la partida se rehizo en la ronda seria buena opcion borrar esa fila ya que no benefician esos datos

-La columna agent tiene un desbalance ya que esta persona tiene prefencias por un agente en especifico, lo que en los video juego se le llama que es su main, ese agente es con lo que el se siente comodo y puede representar una desventaja al acostumbrarse a jugar solo con él y al ser sacado de su zona de conford de jugar, es decir, le toque jugar otro agente esto puede afectar su desempeño ya que cada agente tiene habilidades diferentes, es de ver la manera como se analizan sus partidad tomando en cuenta este factor.

-round_loses y round_wins estan altamente correlacionados con resultado y entre si, ademas que indica que roud_loses tambien esta correlacionado con muertes, esto indica que las partidas que muere mucho este tiende a perderlas 

-rank, act, episodie que es el rango del jugador esta relacionado con el acto y el episodio, cada acto tiene 4 episodios en cada acto se reinicia el rango por eso la alta correlacion, lo que se podria hacer en combinar ambas columnas como por ejemplo, act 1 episodio 1.


-avg_dmg y avg_dmg_delta son estadisticas del daño del jugador la primero por ronda y la segunda el daño promedio por ronda por eso su altacorrelacion entre si se podria evitar eliminar el delta, ya que es un promedio del avg_dmg

-death, kills y kdr estan relacionadas entre si ya que el kdr es una estadistica de cuantas kill hizo por muertes (kills / muertes) el kdr es un dato importante para el analisis de los datos.
